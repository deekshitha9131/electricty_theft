```python
import numpy as np
import matplotlib.pyplot as plt

sample_ids = np.random.choice(df['CONS_NO'], 3, replace=False)

for cid in sample_ids:
    row = df[df['CONS_NO'] == cid].iloc[0]
    ts = row[2:]  # consumption values only
    
    plt.figure(figsize=(12,4))
    plt.plot(ts.values)
    plt.title(f"Consumption Curve â€” Customer {cid}")
    plt.xlabel("Days")
    plt.ylabel("Consumption")
    plt.show()
```

```python
id_cols = ['CONS_NO', 'FLAG']
df_long = df.melt(id_vars=id_cols, var_name='DATE', value_name='CONSUMPTION')
df_long['DATE'] = pd.to_datetime(df_long['DATE'], format='%Y/%m/%d', errors='coerce')
df_long = df_long.dropna(subset=['CONSUMPTION'])  # remove NaNs
df_long.head()
```

```python
df_long = df_long.dropna(subset=['DATE'])
```

```python
df_long.to_csv("../data/processed/cleaned_long.csv", index=False)
```
